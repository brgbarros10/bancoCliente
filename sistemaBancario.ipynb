```python
def menu():
  print('[S] - saque')
  print('[D] - depósito')
  print('[O] - saldo')
  print('[E] - extrato')
  print('[C] - criar conta')
  print('[N] - novo cliente')
  print('[L] - listar contas')
  print('[Q] - sair')
  return input()

def deposito(valor, saldo, extrato, /):
  if valor<0:
    print('Valor inválido, depósito não efetuado!\n')
    return valor
  else:
    print(f'Depósito realizado no valor de R$ {valor:.2f}\n')
    extrato.append(["Depósito", valor])
    saldo += valor
    return saldo, extrato

def saldo(saldo):
  print(f'Teu saldo atual é: R$ {saldo}\n')

def saque(*, valor, saldo, extrato, limite, numero_saque, limite_saque):
  if numero_saque==limite_saque:
    print('Você já realizou os saques permitidos para hoje!\n')
  elif valor>limite:
    print("Valor máximo para esta operação R$ 500,00\n")
  elif saldo<=0:
    print("Saldo indisponível!\n")
  else:
    print(f'Saque realizado no valor de R$ {valor:.2f}\n')
    saldo -= valor
    numero_saque += 1
    extrato.append(['Saque', valor])
  return saldo, extrato, numero_saque

def imprimir_extrato(saldo,/,*,extrato):
  print(f'EXTRATO\n')
  if not extrato:
    print('Não foram realizadas operações nesta conta! \n')
  else:
    for i in extrato:
      print(f'{i[0]} no valor de\t R$ {i[1]:.2f}')
    print(f'Total na conta\t R$ {saldo:.2f}\n')

def criar_cliente(clientes):
  cpf = input('Digite o CPF (só numeros):')

  cliente = busca_cliente(cpf, clientes)
  if cliente:
    print('CPF já cadastrado!\n')
  else:
    nome = input('Digite o nome completo: ')
    data_nascimento = input('Digite a data de nascimento (dd-mm-aaa)')
    endereco = input('Digite o endereço (logradouro, número - bairro - cidade/sigla edtado):')

    clientes.append({'nome':nome, 'data_nascimento':data_nascimento, 'cpf':cpf, 'endereco':endereco})
    print('Cliente cadastrado com sucesso!\n')
    return clientes

def busca_cliente(cpf, clientes):
    cliente = [cliente for cliente in clientes if cliente['cpf']==cpf]
    return cliente[0] if cliente else None

def criar_conta(agencia, numero_conta, clientes):
    cpf = input('informe o cpf do cliente')
    cliente = busca_cliente(cpf, clientes)

    if cliente:
      print('Conta criada com sucesso!\n')
      return {'agencia':agencia, 'numero_conta': numero_conta, 'cliente': cliente}

    print('Cliente não encontrado, conta não criada!\n')
    return None

def listar_contas(contas):
   print('Contas ativas:\n')
   for conta in contas:
    print(conta)


saldo = 0
extrato = []
numero_saque = 0
LIMITE = 500
LIMITE_SAQUE = 3
agencia = '0001'
clientes = []
contas = []
numero_conta = 1

while True:
  opcao = menu()
  opcao = opcao.upper()
  if opcao == 'S':
     valor = float(input('Valor do saque: '))
     saldo, extrato, numero_saque = saque(valor=valor, saldo=saldo, extrato=extrato,
                                    limite=LIMITE, numero_saque=numero_saque, limite_saque=LIMITE_SAQUE)
  elif opcao == 'D':
     valor = float(input('Valor do deposito: '))
     saldo, extrato = deposito(valor, saldo, extrato)

  elif opcao == 'O':
    saldo(saldo)

  elif opcao == 'E':
    imprimir_extrato(saldo, extrato=extrato)

  elif opcao == 'N':
    clientes = criar_cliente(clientes)

  elif opcao == 'C':
    conta = criar_conta(agencia, numero_conta, clientes)
    if conta:
      contas.append(conta)
      numero_conta += 1

  elif opcao == 'L':
    listar_contas(contas)

  else:
    break
```
